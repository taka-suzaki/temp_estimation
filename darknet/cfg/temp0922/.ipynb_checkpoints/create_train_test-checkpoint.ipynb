```python
import os
import random
```

```python
img_dir = './datasets/'
train_txt = 'train.txt'
test_txt = 'test.txt'
```

```python
img_list = [f for f in os.listdir(img_dir) if f.endswith('.jpg')]
print(len(img_list))
```

```python
random.shuffle(img_list)
img_list[:5]
```

```python
pwd = %pwd
train_path =  './' + train_txt
test_path = './' + test_txt
pwd
```

```python
train_num = int(len(img_list) / 10 * 9)
train_num
```

```python
train_list = img_list[0:train_num]
test_list = img_list[train_num:]
print(len(train_list), len(test_list))
```

```python
with open(train_path, 'wt') as f:
    for train in train_list:
        print(train)
        f.write(pwd + '/datasets/' + train + '\n')
        
with open(test_path, 'wt') as f:
    for test in test_list:
        f.write(pwd + '/datasets/' + test + '\n')
```
